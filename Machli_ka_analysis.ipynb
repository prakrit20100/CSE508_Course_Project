```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
new_df = pd.read_csv('merged_df.csv')
```

```python
grouped_data_classification = new_df.groupby('classification')
```

```python
for group_name, group_data in grouped_data_classification:
    print("Group:", group_name)
```

```python
Classes = []
total_data = []
```

```python
for group_name, group_data in grouped_data_classification:
    Classes.append(group_name)
    total_data.append(len(group_data))
```

```python
print('Classes:', Classes)
print('Datas:', total_data)
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
bars = plt.bar(Classes, total_data, color='skyblue')

# Adding values on top of bars
for bar, val in zip(bars, total_data):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), val,
             ha='center', va='bottom')

plt.xlabel('Classes')
plt.ylabel('Datas')
plt.title('Histogram of Classes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_data_endyear = new_df.groupby('endyear')
```

```python
for group_name, group_data in grouped_data_classification:

    print("Group:", group_name)
    print(len( group_data))
```

```python
group_by_era = {'Before 1000':{},
                '1000-1100':{},
                '1100-1200':{},
                '1200-1300':{},
                '1300-1400':{},
                '1400-1500':{},
                '1500-1600':{},
                '1600-1700':{},
                '1700-1800':{},
                '1800-1900':{},
                '1900-2000':{},
                '2000-2100':{}
                }

for year, group_data in grouped_data_endyear:
    y = year
    if y < 1000:
        if y not in group_by_era['Before 1000']:
            group_by_era['Before 1000'][y] = 0
        group_by_era['Before 1000'][y] += len( group_data)
    elif 1000 <= y < 1100:
        if y not in group_by_era['1000-1100']:
            group_by_era['1000-1100'][y] = 0
        group_by_era['1000-1100'][y] += len( group_data)
    elif 1100 <= y < 1200:
        if y not in group_by_era['1100-1200']:
            group_by_era['1100-1200'][y] = 0
        group_by_era['1100-1200'][y] += len( group_data)
    elif 1200 <= y < 1300:
        if y not in group_by_era['1200-1300']:
            group_by_era['1200-1300'][y] = 0
        group_by_era['1200-1300'][y] += len( group_data)
    elif 1300 <= y < 1400:
        if y not in group_by_era['1300-1400']:
            group_by_era['1300-1400'][y] = 0
        group_by_era['1300-1400'][y] += len( group_data)
    elif 1400 <= y < 1500:
        if y not in group_by_era['1400-1500']:
            group_by_era['1400-1500'][y] = 0
        group_by_era['1400-1500'][y] += len( group_data)
    elif 1500 <= y < 1600:
        if y not in group_by_era['1500-1600']:
            group_by_era['1500-1600'][y] = 0
        group_by_era['1500-1600'][y] += len( group_data)
    elif 1600 <= y < 1700:
        if y not in group_by_era['1600-1700']:
            group_by_era['1600-1700'][y] = 0
        group_by_era['1600-1700'][y] += len( group_data)
    elif 1700 <= y < 1800:
        if y not in group_by_era['1700-1800']:
            group_by_era['1700-1800'][y] = 0
        group_by_era['1700-1800'][y] += len( group_data)
    elif 1800 <= y < 1900:
        if y not in group_by_era['1800-1900']:
            group_by_era['1800-1900'][y] = 0
        group_by_era['1800-1900'][y] += len( group_data)
    elif 1900 <= y < 2000:
        if y not in group_by_era['1900-2000']:
            group_by_era['1900-2000'][y] = 0
        group_by_era['1900-2000'][y] += len( group_data)
    elif 2000 <= y < 2100:
        if y not in group_by_era['2000-2100']:
            group_by_era['2000-2100'][y] = 0
        group_by_era['2000-2100'][y] += len( group_data)
```

```python
total = 0
for i in group_by_era:
    y = group_by_era[i]
    for j in y:
        total += y[j]
```

```python
# Extracting data from the dictionary
eras = list(group_by_era.keys())
counts = [sum(group_by_era[era].values()) for era in eras]

# Plotting the histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(eras, counts, color='skyblue')

# Adding counts above each bar
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10, str(count), ha='center', va='bottom')

plt.xlabel('Era')
plt.ylabel('Count')
plt.title('Histogram of Data by Era')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
